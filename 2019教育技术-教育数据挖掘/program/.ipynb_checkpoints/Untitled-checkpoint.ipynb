```python
#如何利用pandas进行数据分组，cut这样的一个函数data['G3']表示读取的数据列，bins划分的界限，labels数据分列后的标记， 
#include_lowest=True表示更新的数据也在原始的数据集中进行更新
listBins = [0,9,20]
listLabels = ['FAIL','PASS']
data['Binary-classification'] = pd.cut(data['G3'], bins=listBins, labels=listLabels, include_lowest=True)
```

## 三、预处理 

### （一）将所有的object类型转化为数值型 

```python
df2 = data.select_dtypes('object').copy()#将数据中所有的文本全部提取出来，复制进df2中
df2.head()
heads=df2.columns.values#获得全部的列名
print(heads)
```

因为我们已经知道，这里面数据有两种，第一种是数值（int），第二种是字符（object类型）---->1-of-C

**思路**
- 1.读取所有位object类型的数据出来
    pandas里面有一个函数，这个函数是专门获取某一种类型列数据的，select_dtypes(),里面的参数是检索的的数据类型，df2中间承载了所有字符型数据列
    通过pandas里面的columns的属性得到，定位到，然后通过value属性获取列名
- 2.将object里面的文本自动编码成数值
    对于如schoolsup这一类型的，数据，因为含有yes和no是相同的数据类型，所以我们可以批量处理掉
    replace利用到了replace函数，该函数在pandas里面就用与处理，列值的替换

```python
for col in heads[-8:]:#对yes和no进行批量化操作
    df2.replace(['yes','no'],[1,0],inplace=True)
```

```python
df2
```

```python
for col in heads[0:9]:#其余的进行转化
    strs = df2[col].value_counts()#value_counts()表示对列字符进行统计
    print(strs)
    value_map = dict((v, i) for i,v in enumerate(strs.index))
    print(value_map)
    df2.replace({col:value_map},inplace=True)
```

```python
df2#df2中间已经存储了全部的字符型转化为数值新的数据了吧，
```

```python
df3 = data.select_dtypes('int64').copy()#将数值型的数据提取出来，并且完成拼接
```

```python
preprodata = pd.concat([df2,df3],axis=1)#合并数值化以后的数据concat函数拼接数据
preprodata
```

### （二） 数据标准化

P5:all attributes were standardized to a zero mean and one standard deviation

```python
#标签与特征数据拆分
G1_data=preprodata['G1']#G1
G2_data=preprodata['G2']#G2成绩，某种方式上来说，他们代表着类型
G3_data=preprodata['G3']#G3某种方式上说，他们是最后的结果，是类型
X_data=preprodata.drop(['G1','G2','G3'],axis=1)#drop表示剔除某些列
```

```python
from sklearn.preprocessing import StandardScaler #导入标注化处理包
ss_X=StandardScaler()                            #初始化特征的标准化器
X_data=ss_X.fit_transform(X_data)                #对训练和测试数据的特征进行标准化处理

print(X_data)#此时的X_data其实是经过标准化处理的了
```

## 三、模型训练 

### （一）导入算法包 

```python
from sklearn.linear_model import LogisticRegression #导入逻辑回归算法
from sklearn.model_selection import cross_val_score # 导入交叉验证库
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

### （二） 二分类情况（Binary Classification）

```python
#加载出二分类数据
BiLabel = data['Binary-classification']
BiLabel.replace(['FAIL','PASS'],[0,1],inplace=True)
BiLabel
```

#### 1.逻辑回归算法拟合模型

```python
x_data = X_data
lr=LogisticRegression()
score=cross_val_score(lr,x_data,BiLabel,cv=10,scoring='accuracy')#10折：交叉验证

print ('平均准确率',score.mean())
print ('每则准确率',score)
```

#### 2.朴素贝叶斯算法拟合模型 

```python
from sklearn.naive_bayes import GaussianNB
x_data = X_data
nb=GaussianNB()
score=cross_val_score(nb,x_data,BiLabel,cv=10,scoring='accuracy')#10折：交叉验证

print ('平均准确率',score.mean())
print ('每则准确率',score)
```

#### 3.支持向量机算法拟合模型 

```python
from sklearn.svm import SVC
x_data = X_data
svc=SVC(kernel='linear', C=1)
score=cross_val_score(svc,x_data,BiLabel,cv=10,scoring='accuracy')#10折：交叉验证

print ('平均准确率',score.mean())
print ('每则准确率',score)
```

#### 4.决策树算法拟合模型 

```python
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier

x_data = X_data
tree = DecisionTreeClassifier(random_state=0)
score=cross_val_score(tree,x_data,BiLabel,cv=10,scoring='accuracy')#10折：交叉验证

print ('平均准确率',score.mean())
print ('每则准确率',score)
```
